```python
def checkDataset(data):
    datadict = {}
    for row in data: 
        changeId = row[2]
        fileName = row[1]
        label = row[0]
        if changeId in datadict:
            if fileName in datadict[changeId]:
                datadict[changeId][fileName].append(label)
            else:
                datadict[changeId][fileName] = [label]
        else:
            datadict[changeId] = {fileName:[]}
    avg = []
    label_avg = []
    for changeId, value in datadict.items():
        for fileName, labels in value.items():
            pos_labels = 0
            avg.append(len(labels))
            for label in labels:
                if label == 1:
                    pos_labels += 1
            if(len(labels) == 0):
                label_avg.append(0)
            else:
                label_avg.append(pos_labels/len(labels))
    print("avg LOC",np.mean(avg))
    print("avg LOC",np.mean(label_avg))
```

```python
def getDatasetFromRawData(data, bias):
    row_data = data[0:,3]
    row_data_Y = data[0:,0]
    row_data_deletions = data[0:,6]
    row_data_additions = data[0:,7]
    row_data_changedLine = data[0:,8]
  
    Y_train = []
    is_comment = 0
    not_comment = 0
    for element in row_data_Y:
        if(element == 0):
            Y_train.append(False)
            not_comment += 1
        else:
            Y_train.append(True)
            is_comment += 1
    Y_train = np.array(Y_train)

    first = int(len(data)*0.6)
    # second = int(len(data)*0.2) + first + 49
    divider = int(len(data)*0.2) + first + bias
#     second = int(len(data)*0.2) + first + 2
    # second = int(len(data)*0.7)#6:4

    # data_count_vect = CountVectorizer(lowercase=False)
    # data_count_vect = CountVectorizer(lowercase=False)
    data_count_vect = CountVectorizer(min_df=2, max_df=0.5)
    # data_count_vect = CountVectorizer(min_df=3, max_df=0.5)

    train_row_data = row_data[:divider]
    test_row_data = row_data[divider:]

    train_row_data_deletions = row_data_deletions[:divider]
    test_row_data_deletions = row_data_deletions[divider:]

    train_row_data_additions = row_data_additions[:divider]
    test_row_data_additions = row_data_additions[divider:]

    train_row_data_changedLine = row_data_changedLine[:divider]
    test_row_data_changedLine = row_data_changedLine[divider:]

    data_train_counts = data_count_vect.fit_transform(train_row_data)
    data_test_counts = data_count_vect.transform(test_row_data)

    final_train_X = np.hstack((data_train_counts.toarray(),train_row_data_deletions[:,None]))
    final_train_X = np.hstack((final_train_X,train_row_data_additions[:,None]))
    final_train_X = np.hstack((final_train_X,train_row_data_changedLine[:,None]))

    final_test_X = np.hstack((data_test_counts.toarray(),test_row_data_deletions[:,None]))
    final_test_X = np.hstack((final_test_X,test_row_data_additions[:,None]))
    final_test_X = np.hstack((final_test_X,test_row_data_changedLine[:,None]))
    final_train_y = Y_train[:divider]
    final_test_y = Y_train[divider:]
    print(data[:divider][-1])
    print(data[divider:][0])
    print(len(final_train_X),len(final_test_X))
    del data_train_counts,data_test_counts,train_row_data,test_row_data,data,row_data,row_data_Y 
    return final_train_X,final_train_y,final_test_X,final_test_y,divider,data_count_vect
```

```python
def dataStatastic(train_X, train_y, test_X, test_y,data_count_vect):
    train_true_label = 0
    test_true_label = 0
    for label in train_y:
        if label == 1:
            train_true_label += 1
    for label in test_y:
        if label == 1:
            test_true_label += 1
    print("training dataset size:", len(train_X))
    print("training dataset positive label size:", train_true_label, train_true_label/len(train_y))
    print("test dataset size:", len(test_X))
    print("test dataset positive label size:", test_true_label, test_true_label/len(test_X))
    print("feature_size:", len(data_count_vect.get_feature_names()))
```

```python
def printResult(x, y, model):
    print(roc_auc_score(y, model.predict_proba(x)[:,1]))
    print(precision_score(y, model.predict(x)))
    print(recall_score(y, model.predict(x)))
    print(f1_score(y, model.predict(x)))
    print(matthews_corrcoef(y, model.predict(x)))
    print(confusion_matrix(y, model.predict(x)))
```

```python
def trainRFmodel(project,rf_train_X,rf_train_y,rf_test_X,rf_test_y,seed):
    print("RF seed:" + str(seed))
    train_rf_model_path = model_path+'/RQ2_2_rf_'+project+'-'+str(seed)+'.pkl'
    print(train_rf_model_path)
    if not os.path.exists(train_rf_model_path):
        rf = RandomForestClassifier(n_estimators=200,n_jobs=-1,random_state=seed)
        rf_X, rf_y = SMOTE(k_neighbors=10, random_state=seed).fit_resample(rf_train_X, rf_train_y)
        rf.fit(rf_X,rf_y)
        rf_ouput = open(train_rf_model_path, 'wb')
        pickle.dump(rf,rf_ouput)
        print("finish to creat a new model")
    else:
        with open(train_rf_model_path,'rb') as f:
            rf = pickle.load(f)
    printResult(rf_test_X,rf_test_y,rf)
    return rf

# TN FP
# FN TP
```

```python
def getCorrectedPredictFileDict(model,data,test_X,test_y,divider):
    predict_counter = 0;
    correct_prediction_dict = {}
    predic_set = []
    test_dataset_row = data[divider:]
    predict_label = model.predict(test_X)
    for index in range(len(predict_label)):
        if(predict_label[index] == test_y[index] and test_y[index] == 1):
            predict_counter += 1
            changeId = test_dataset_row[index][2]
            fileName = test_dataset_row[index][1]
            if changeId in correct_prediction_dict:
                correct_prediction_dict[changeId][fileName] = test_X[index]
            else:
                correct_prediction_dict[changeId] = {fileName:test_X[index]}
            predic_set.append((changeId,fileName))
    return correct_prediction_dict,set(predic_set)
```

```python
def getCorrectedPredictLineDict(correct_prediction_dict,line_data, ngram=False):
    line_check_counter = 0
    line_predict_dict = {}
    line_predict_set = []
    for line_element in line_data:
        line_predict_fileName = line_element[1]
        line_predict_changeId = line_element[2]
        if ngram == True:
            line_predict_code = line_element[3].replace("<NUMBER>","")
        else:
            line_predict_code = line_element[3]
        line_predict_loc = line_element[10]
        line_predict_label = line_element[0]
        if line_predict_changeId in correct_prediction_dict:
            if line_predict_fileName in correct_prediction_dict[line_predict_changeId]:
                if line_predict_changeId in line_predict_dict:
                    if line_predict_fileName in line_predict_dict[line_predict_changeId]:
                        line_predict_dict[line_predict_changeId][line_predict_fileName].append((line_predict_code, line_predict_loc, line_predict_label))
                    else:
                        line_predict_dict[line_predict_changeId][line_predict_fileName] = [(line_predict_code, line_predict_loc, line_predict_label)]
                else:
                    line_predict_dict[line_predict_changeId] = {line_predict_fileName:[(line_predict_code, line_predict_loc, line_predict_label)]}
                line_predict_set.append((line_predict_changeId,line_predict_fileName))
    return line_predict_dict,set(line_predict_set)
```

```python
def getLimeExplainerCountDict(project,feature_dict):
    count_dict = {}
    stop_tokens = ["*deletedLine","*addedLine","*changedLine","i", "me", "my", "myself", "we", "our", "ours", "ourselves", "you", "your", "yours", "yourself", "yourselves", "he", "him", "his", "himself", "she", "her", "hers", "herself", "it", "its", "itself", "they", "them", "their", "theirs", "themselves", "what", "which", "who", "whom", "this", "that", "these", "those", "am", "is", "are", "was", "were", "be", "been", "being", "have", "has", "had", "having", "do", "does", "did", "doing", "a", "an", "the", "and", "but", "if", "or", "because", "as", "until", "while", "of", "at", "by", "for", "with", "about", "against", "between", "into", "through", "during", "before", "after", "above", "below", "to", "from", "up", "down", "in", "out", "on", "off", "over", "under", "again", "further", "then", "once", "here", "there", "when", "where", "why", "how", "all", "any", "both", "each", "few", "more", "most", "other", "some", "such", "no", "nor", "not", "only", "own", "same", "so", "than", "too", "very", "s", "t", "can", "will", "just", "don", "should", "now"]
    python_common_tokens = ['abs','delattr','hash','memoryview','set','all','dict','help','min','setattr','any','dir','hex','next','slice','ascii','divmod','id','object','sorted','bin','enumerate','input','oct','staticmethod','bool','eval','int','open','str','breakpoint','exec','isinstance','ord','sum','bytearray','filter','issubclass','pow','super','bytes','float','iter','print','tuple','callable','format','len','property','type','chr','frozenset','list','range','vars','classmethod','getattr','locals','repr','zip','compile','globals','map','reversed','__import__','complex','hasattr','max','round','False','await','else','import','passNone','break','except','in','raise','True','class','finally','is','return','and','continue','for','lambda','try','as','def','from','nonlocal','while','assert','del','global','not','with','async','elif','if','or','yield', 'self']
    c_common_tokens = ['auto','const','double','float','int','short','struct','unsigned','break','continue','else','for','long','signed','switch','void','case','default','enum','goto','register','sizeof','typedef','volatile','char','do','extern','if','return','static','union','while','asm','namespace','try','bool','explicit','new','typeid','catch','false','operator','template','typename','class','friend','private','this','using','inline','public','throw','virtual','delete','mutable','protected','true']
    if project == 'nova' or project == 'ironic':
        print("using python stopword")
        common_tokens = python_common_tokens
    if project == 'base':
        print("using c stopword")
        common_tokens = c_common_tokens
    for key,value in feature_dict.items():
        for fileName, score_list in value.items():
            important_tokens = [token for token in score_list.keys() if token not in stop_tokens and token not in common_tokens]
            important_tokens = important_tokens[:int(len(important_tokens))]
            if key in count_dict:
                count_dict[key][fileName] = important_tokens
            else:
                count_dict[key] = {fileName: important_tokens}
    return count_dict
```

```python
def evalLine(line_predict_dict,feature_dict, length_count_dict):
    eval_dict = {}
    eval_index = 0
    for key,value in line_predict_dict.items():
        for fileName, locs in value.items():
            for loc in locs:
                loc_score_eval = 0
#                 loc_code_eval = loc[0].lower()
                loc_raw_code_eval = loc[0]
                loc_code_eval = " ".join(set(loc[0].lower().split()))
                loc_line_eval = loc[1]
                loc_label_eval = loc[2]
                for token in loc_code_eval.split():
                    loc_score_eval += feature_dict[key][fileName].get(token,0)
#                 loc_score_eval = (loc_score_eval/len(loc_code_eval.split()))*length_count_dict.get(len(loc_raw_code_eval.split()),0)
                if loc_score_eval == 0:
                    loc_score_eval = 0.0000000001
                loc_score_eval = (loc_score_eval)*length_count_dict.get(len(loc_raw_code_eval.split()),0)
#                 loc_score_eval = (loc_score_eval/len(loc_code_eval.split()))
                if key in eval_dict:
                    if fileName in eval_dict[key]:
                        eval_dict[key][fileName].append((loc_code_eval,loc_line_eval,loc_label_eval,loc_score_eval))
                    else:
                        eval_dict[key][fileName] = [(loc_code_eval,loc_line_eval,loc_label_eval,loc_score_eval)]
                else:
                    eval_dict[key] = {fileName:[(loc_code_eval,loc_line_eval,loc_label_eval,loc_score_eval)]}
    return eval_dict
```

```python
def evalLineByCount(line_predict_dict,count_dict,length_count_dict):
    eval_count_dict = {}
    eval_index = 0
    for key,value in line_predict_dict.items():
        for fileName, locs in value.items():
            for loc in locs:
                loc_count_score_eval = 0
                loc_count_eval = 0
                loc_code_eval = " ".join(set(loc[0].lower().split()))
                loc_line_eval = loc[1]
                loc_label_eval = loc[2]
                for token in loc_code_eval.split():
                    if(token in count_dict[key][fileName]):
                        loc_count_eval += 1
                loc_count_score_eval = loc_count_eval/len(loc_code_eval.split())
#                 loc_count_score_eval = loc_count_score_eval * length_count_dict.get(len(loc[0].split()),0)
                loc_count_score_eval = loc_count_eval
                if key in eval_count_dict:
                    if fileName in eval_count_dict[key]:
                        eval_count_dict[key][fileName].append((loc_code_eval,loc_line_eval,loc_label_eval,loc_count_score_eval,loc_count_eval))
                    else:
                        eval_count_dict[key][fileName] = [(loc_code_eval,loc_line_eval,loc_label_eval,loc_count_score_eval,loc_count_eval)]
                else:
                    eval_count_dict[key] = {fileName:[(loc_code_eval,loc_line_eval,loc_label_eval,loc_count_score_eval,loc_count_eval)]}
    return eval_count_dict
```

```python
def evalLineNgram(project,seed):
    data_entropy = pd.read_csv(eval_file_path + '/entropy_revised_'+project+'-'+str(seed)+'.csv', dtype=None, sep=',',header=None).to_numpy()
    data_ngram_test = pd.read_csv(eval_file_path + '/lime_test_'+project+'-'+str(seed)+'.csv', dtype=None, sep=',',header=None).to_numpy()

    ngram_eval_line = {}
    for index in range(len(data_ngram_test)):
        eval_line_ngram_changeId = data_ngram_test[index][0]
        eval_line_ngram_fileName = data_ngram_test[index][1]
        eval_line_ngram_code = data_ngram_test[index][2]
        eval_line_ngram_line = data_ngram_test[index][3]
        eval_line_ngram_label = data_ngram_test[index][4]
        eval_line_entropy = data_entropy[index][0]
        if eval_line_ngram_changeId in ngram_eval_line:
            if eval_line_ngram_fileName in ngram_eval_line[eval_line_ngram_changeId]:
                ngram_eval_line[eval_line_ngram_changeId][eval_line_ngram_fileName].append((eval_line_ngram_code,eval_line_ngram_line,eval_line_ngram_label,eval_line_entropy))
            else:
                ngram_eval_line[eval_line_ngram_changeId][eval_line_ngram_fileName] = [(eval_line_ngram_code,eval_line_ngram_line,eval_line_ngram_label,eval_line_entropy)]
        else:
            ngram_eval_line[eval_line_ngram_changeId] = {eval_line_ngram_fileName:[(eval_line_ngram_code,eval_line_ngram_line,eval_line_ngram_label,eval_line_entropy)]}
    return ngram_eval_line
```

```python
def outputSlpTestFile(project,line_predict_dict_ngram,seed):
    csv_ngram_path = eval_file_path + '/ngram_test_revised_'+ project +'-'+str(seed)+'.csv'
    csv_lime_path = eval_file_path + '/lime_test_'+ project +'-'+str(seed)+'.csv'
    csv_ngram_file = open(csv_ngram_path,"w")
    csv_lime_file = open(csv_lime_path,"w")
    csv_ngram_writer = csv.writer(csv_ngram_file,quoting=csv.QUOTE_NONE,escapechar='') 
    csv_lime_writer = csv.writer(csv_lime_file) 
    for key,value in line_predict_dict_ngram.items():
        for fileName, codeLines in value.items():
            for codeLine in codeLines:
                csv_ngram_writer.writerow([codeLine[0]])
                csv_lime_writer.writerow([key, fileName, codeLine[0].strip(),codeLine[1],codeLine[2]])
    csv_ngram_file.close()  
    csv_lime_file.close()
    print("write test datset done")
```

```python
def outputSlpTrainFile(project,seed):
    capitalied_project = project.capitalize()
    data_ngram_train = pd.read_csv(eval_file_path + '/ngramTrain' + capitalied_project + '.csv', dtype=None, sep=',',header=None).to_numpy().tolist()
    data_train_ngram_test = pd.read_csv(eval_file_path + '/ngram_test_revised_' + project + '-'+str(seed)+'.csv', dtype=None, sep=',',header=None).to_numpy().tolist()
    data_ngram_train = [x[0] for x in data_ngram_train]
    data_train_ngram_test = [x[0].strip() for x in data_train_ngram_test]
    new_data_train_ngram_test = list(set(data_ngram_train) - set(data_train_ngram_test))

    csv_ngram_train_path = eval_file_path + '/ngram_train_revised_'+ project +'-'+str(seed)+'.csv'
    csv_new_train_ngram_file = open(csv_ngram_train_path,"w")
    csv_ngram_train_writer = csv.writer(csv_new_train_ngram_file,quoting=csv.QUOTE_NONE,escapechar='') 
    for code in new_data_train_ngram_test:
        csv_ngram_train_writer.writerow([code])
    csv_new_train_ngram_file.close()  
    print("write train datset done")
```

```python
def getCommentedLineInTrain(row_data, line_row_data,divider):
    train_row_data_contents = row_data[:divider]
    train_row_data_dict = {}
    #Build a dict only includes the training data.
    for train_row_data_content in train_row_data_contents:
        train_row_data_label = train_row_data_content[0]
        train_row_data_file = train_row_data_content[1]
        train_row_data_changeId = train_row_data_content[2]
        if train_row_data_label == 1:
            if train_row_data_changeId in train_row_data_dict:
                train_row_data_dict[train_row_data_changeId].append(train_row_data_file)
            else:
                train_row_data_dict[train_row_data_changeId] = [train_row_data_file]
    line_row_data_dict = {}
    code_length_distribution = []
    for line_code_content in line_row_data:
        train_row_line_data_label = line_code_content[0]
        train_row_line_data_file = line_code_content[1]
        train_row_line_data_changeId = line_code_content[2]
        train_row_line_data_code = line_code_content[3]
        #check if the LOC's file and change id in the training dataset 
        if train_row_line_data_changeId in train_row_data_dict and train_row_line_data_file in train_row_data_dict[train_row_line_data_changeId] and train_row_line_data_label == 1:
            #Save to a new dict
            code_length_distribution.append(len(train_row_line_data_code.split()))
            if train_row_line_data_changeId in line_row_data_dict:
                if train_row_line_data_file in line_row_data_dict[train_row_line_data_changeId]:
                    line_row_data_dict[train_row_line_data_changeId][train_row_line_data_file].append(train_row_line_data_code)
                else:
                    line_row_data_dict[train_row_line_data_changeId] = {train_row_line_data_file:[train_row_line_data_code]}
            else:
                line_row_data_dict[train_row_line_data_changeId] = {train_row_line_data_file:[train_row_line_data_code]}
    length_count_dict = {}
    (unique, counts) = np.unique(code_length_distribution,return_counts=True)
    for index in range(len(unique)):
        length_count_dict[unique[index]] = float(counts[index]/len(code_length_distribution))
    return length_count_dict,code_length_distribution,line_row_data_dict,train_row_data_dict
```

```python
def getLineStatistic(line_predict_dict):
#     print(line_predict_dict)
    total = 0
    positive = 0
    for key,value in line_predict_dict.items():
        for fileName,codes in value.items():
            for code in codes:
                total += 1
                if code[2] == 1:
                    positive += 1
    print("line statistic total:", total)
    print("line statistic positive:", positive)
    print("line statistic %:", positive/total)
```

```python
#Top k precision
def getTopKPrecision(k,sorted_eval_dict):
    k_precision = []
    for key,value in sorted_eval_dict.items():
        for fileName,locs in value.items():
            commented_locs_amount = 0
            total_commented_locs_amount = 0
            if k >= len(locs):
                k = len(locs)
            for loc in locs[:k]:
                if loc[2] == 1:
                    commented_locs_amount += 1
            for loc in locs:
                if loc[2] == 1:
                    total_commented_locs_amount += 1
            k_precision.append(commented_locs_amount/total_commented_locs_amount)
#     return sorted(k_precision, reverse=True)
    return k_precision
```

```python
#% of files have at least one inline comment 
def getPercentageTopK(k,sorted_eval_dict):
    commented_file_number = 0
    total_file_number = 0
    for key,value in sorted_eval_dict.items():
        for fileName,locs in value.items():
            commented_file = False
            if k >= len(locs):
                k = len(locs)
            for loc in locs[:k]:
                if loc[2] == 1:
                    commented_file = True
            if commented_file:
                commented_file_number += 1
            total_file_number += 1        
    return commented_file_number/total_file_number
```

```python
def getIFA(eval_dict):
    result_ifa = []
    total_files = 0
    sorted_eval_dict = {}
    ifa_result_dict = {}
    for key, value in eval_dict.items():
        for fileName, info in value.items():
            sorted_info = sorted(info, key=lambda tup:tup[3],reverse=True)
            label_position = 1
            total_files += 1
            for tupe in sorted_info:
                if tupe[2] == 1:
                    break;
                else:
                    label_position += 1
            if key in sorted_eval_dict:    
                sorted_eval_dict[key][fileName] = sorted_info
            else:
                sorted_eval_dict[key] = {fileName:sorted_info}
#             print(key, fileName,label_position)
            result_ifa.append(label_position - 1)
            if key in ifa_result_dict:
                ifa_result_dict[key][fileName] = label_position - 1
            else:
                ifa_result_dict[key] = {fileName:label_position - 1}
    print(np.median(result_ifa))
    return result_ifa,sorted_eval_dict,ifa_result_dict
```

```python
#Recall@20%LOC
def getRecallLoc(percentage,sorted_eval_dict):
    recall_20_loc = []
    recall_20_loc_number = []
    for key,value in sorted_eval_dict.items():
        for fileName,locs in value.items():
            locs_20_index = round(len(locs) * percentage)
            recall_20_loc_number.append(locs_20_index + 1)
            commented_locs_amount = 0
            commented_locs_amount_20 = 0
            for loc in locs:
                if loc[2] == 1:
                    commented_locs_amount += 1
            for loc in locs[:locs_20_index]:
                if loc[2] == 1:
                    commented_locs_amount_20 += 1
            recall_20_loc.append(float(commented_locs_amount_20/commented_locs_amount)) 
#             print(locs,len(locs),len(locs[:locs_20_index]),commented_locs_amount,commented_locs_amount_20)
#     return sorted(recall_20_loc, reverse=True),recall_20_loc_number
    return recall_20_loc,recall_20_loc_number
```

```python
#Effort@20%Recall
def getEffortRecall(percentage,sorted_eval_dict):
    effort_20_recall = []
    for key,value in sorted_eval_dict.items():
        for fileName,locs in value.items():
            commented_locs = []
            for loc in locs:
                if loc[2] == 1:
                    commented_locs.append(loc)
            target_commented_index = round(len(commented_locs) * percentage)
            if target_commented_index == 0:
                target_commented_index = 0
            if target_commented_index >= len(commented_locs):
                target_commented_index = len(commented_locs) - 1

            target_20_loc_index = locs.index(commented_locs[target_commented_index])
            effort_20_recall.append(target_20_loc_index/len(locs))
#     return sorted(effort_20_recall)
    return effort_20_recall
```

```python
#MRR
def getMRR(sorted_eval_dict):
    mrr = []
    mrr_dict = {}
    for key,value in sorted_eval_dict.items():
        for fileName,locs in value.items():
            total_mrr = 0
            total_comment = 0;
            for index in range(len(locs)):
                if locs[index][2] == 1:
                    total_mrr += 1/(index+1)
                    total_comment += 1
            if key in mrr_dict:
                if fileName in mrr_dict[key]:
                     mrr_dict[key][fileName] = (key,fileName,total_mrr/total_comment)
                else:
                    mrr_dict[key] = {fileName:(key,fileName,total_mrr/total_comment)}
            else:
                mrr_dict[key] = {fileName:(key,fileName,total_mrr/total_comment)}
            mrr.append(total_mrr/total_comment)
#     return sorted(mrr),mrr_dict
    return mrr,mrr_dict
```

```python
#MAP
def getMAP(sorted_eval_dict):
    ap_list = []
    ap_dict = {}
    for key,value in sorted_eval_dict.items():
        for fileName,locs in value.items():
            total_ap = 0
            total_comment = 0;
            for index in range(len(locs)):
                if locs[index][2] == 1:
                    total_comment += 1
                    total_ap += total_comment/(index+1)              
            if key in ap_dict:
                if fileName in ap_dict[key]:
                     ap_dict[key][fileName] = (key,fileName,total_ap/total_comment)
                else:
                    ap_dict[key] = {fileName:(key,fileName,total_ap/total_comment)}
            else:
                ap_dict[key] = {fileName:(key,fileName,total_ap/total_comment)}
            ap_list.append(total_ap/total_comment)
    return ap_list,np.mean(ap_list),ap_dict
```

```python
#Summation
def summationResult(seed,projectName,data,line_data,bias):
    projectName = projectName
    #RF
    train_X, train_y, test_X, test_y,divider,data_count_vect = getDatasetFromRawData(data, bias)
    dataStatastic(train_X, train_y, test_X, test_y,data_count_vect)
    rf = trainRFmodel(projectName,train_X, train_y, test_X, test_y,seed)
    correct_prediction_dict, predic_set = getCorrectedPredictFileDict(rf,data,test_X,test_y,divider)
    line_predict_dict,line_predict_set = getCorrectedPredictLineDict(correct_prediction_dict,line_data)
    feature_dict = getLimeExplainerScoreDict(projectName,correct_prediction_dict,data_count_vect,train_X,rf,seed)

    #ngram
    length_count_dict,code_length_distribution,line_row_data_dict,train_row_data_dict = getCommentedLineInTrain(data,line_data,divider)
    line_predict_dict_ngram,line_predict_set_ngram = getCorrectedPredictLineDict(correct_prediction_dict,line_data, ngram=True)


    # #generate files for SLP
    outputSlpTestFile(projectName,line_predict_dict_ngram,seed)
    outputSlpTrainFile(projectName,seed)

    eval_dict = evalLine(line_predict_dict,feature_dict,length_count_dict)
    eval_dict_ngram = evalLineNgram(projectName,seed)

    #IFA
    result_ifa,sorted_eval_dict,ifa_result_dict = getIFA(eval_dict)
    result_ifa_ngram,sorted_eval_dict_ngram,ifa_result_dict_ngram = getIFA(eval_dict_ngram)
    result_top_k = getTopKPrecision(10,sorted_eval_dict)
    result_top_k_ngram = getTopKPrecision(10,sorted_eval_dict_ngram)
#     result_recall_20_loc = getRecallLoc(1,sorted_eval_dict)
#     result_recall_20_loc_ngram = getRecallLoc(1,sorted_eval_dict_ngram)
#     result_effort_20_recall = getEffortRecall(1,sorted_eval_dict)
    result_recall_20_loc,recall_20_loc_number = getRecallLoc(0.2,sorted_eval_dict)
    result_recall_20_loc_ngram,recall_20_loc_number_ngram = getRecallLoc(0.2,sorted_eval_dict_ngram)
    result_effort_20_recall = getEffortRecall(0.2,sorted_eval_dict)
    result_effort_20_recall_ngram = getEffortRecall(0.2,sorted_eval_dict_ngram)
    
    result_mrr,result_mrr_dict = getMRR(sorted_eval_dict)
    result_mrr_ngram,result_mrr_ngram_dict = getMRR(sorted_eval_dict_ngram)
    result_map, result_ap_dict = getMAP(sorted_eval_dict)
    result_map_ngram, result_ap_ngram_dict = getMAP(sorted_eval_dict_ngram)
    result_percentage_top_k = getPercentageTopK(10,sorted_eval_dict)
    result_percentage_top_k_ngram = getPercentageTopK(10,sorted_eval_dict_ngram)
    #ShowFig
    showFigure(result_ifa,result_ifa_ngram, "IFA")
    showSingleFigure(recall_20_loc_number, "How many loc 20%")
    show20effortFigure(result_top_k,result_top_k_ngram, "Top 10 precision")
    show20effortFigure(result_recall_20_loc,result_recall_20_loc_ngram, "Recall 20% LOC")
    show20effortFigure(result_effort_20_recall,result_effort_20_recall_ngram, "Effort 20% Recall")
    show20effortFigure(result_mrr,result_mrr_ngram, "MRR")
    print("MAP",result_map,result_map_ngram)
    print("top-10 percentage of file have at least one in-line comment",result_percentage_top_k,result_percentage_top_k_ngram)
```

```python
def getFileNameList(eval_dict):
    fileNameList = []
    for changeId, value in eval_dict.items():
        for fileName, element in value.items():
            fileNameList.append(fileName)
    return fileNameList
```

```python
def generateResultList(result,approach,project,fileNameList,result_ifa,result_top_k,result_recall_20_loc,result_effort_20_recall,result_mrr,result_map_list):
    for index, element in enumerate(fileNameList):
        result.append({'Technique':approach, 'Datasets':project, 'FileName':element, 'Measure':'Top10Accuracy(Line)', 'Value':result_top_k[index]})
        result.append({'Technique':approach, 'Datasets':project, 'FileName':element, 'Measure':'IFA', 'Value':result_ifa[index]})
        result.append({'Technique':approach, 'Datasets':project, 'FileName':element, 'Measure':'recall20%Loc', 'Value':result_recall_20_loc[index]})
        result.append({'Technique':approach, 'Datasets':project, 'FileName':element, 'Measure':'effort20%Recall', 'Value':result_effort_20_recall[index]})
        result.append({'Technique':approach, 'Datasets':project, 'FileName':element, 'Measure':'MRR', 'Value':result_mrr[index]})
        result.append({'Technique':approach, 'Datasets':project, 'FileName':element, 'Measure':'MAP', 'Value':result_map_list[index]})
    return result
```

```python
def generateRawDataCsv(result,project,eval_dict_sum,eval_dict,eval_dict_ngram):
    for changeId, pair in eval_dict_ngram.items():
        for fileName, value in pair.items():
            for index, element in enumerate(value):
                result.append({"dataset": project, "changeId":changeId, "fileName": fileName, "lineNumber":element[1], "token":element[0],"ngramScore":str(element[3]), "limeScoreSummation":str(eval_dict_sum[changeId][fileName][index][3]),"limeScoreCount":str(eval_dict[changeId][fileName][index][3]), "groundTruth":element[2]})
    return result
```

```python
def generateLimeScoreDataCsv(result,project,feature_dict,line_predict_dict):
    for changeId, pair in line_predict_dict.items():
        for fileName, value in pair.items():
            for index, element in enumerate(value):
                code = list(set(element[0].lower().split()))
                for token in code:
                    token_score = feature_dict[changeId][fileName].get(token,0)
                    result.append({"dataset": project, "changeId":changeId, "fileName": fileName, "lineNumber":element[1],"token":token,"limeScore":token_score})
    return result
```

```python
#count
def countResult(seed,projectName,data,line_data,bias):
    projectName = projectName
    #RF
    train_X, train_y, test_X, test_y,divider,data_count_vect = getDatasetFromRawData(data, bias)
    rf = trainRFmodel(projectName,train_X, train_y, test_X, test_y,seed)
    correct_prediction_dict, predic_set = getCorrectedPredictFileDict(rf,data,test_X,test_y,divider)
    line_predict_dict,line_predict_set = getCorrectedPredictLineDict(correct_prediction_dict,line_data)
    getLineStatistic(line_predict_dict)
#     feature_dict = getLimeExplainerScoreDict(projectName,correct_prediction_dict,data_count_vect,train_X,rf,seed)
    feature_dict = getLimeExplainerScoreDictWithoutExclusion(projectName,correct_prediction_dict,data_count_vect,train_X,rf,seed)
    count_dict = getLimeExplainerCountDict(projectName,feature_dict)

    #ngram
    length_count_dict,code_length_distribution,line_row_data_dict,train_row_data_dict = getCommentedLineInTrain(data,line_data,divider)
    line_predict_dict_ngram,line_predict_set_ngram = getCorrectedPredictLineDict(correct_prediction_dict,line_data, ngram=True)


    # #generate files for SLP
#     outputSlpTestFile(projectName,line_predict_dict_ngram,seed)
#     outputSlpTrainFile(projectName,seed)

    eval_dict_sum = evalLine(line_predict_dict,feature_dict,length_count_dict)
#     eval_dict = evalLine(line_predict_dict,feature_dict,length_count_dict)
    eval_dict = evalLineByCount(line_predict_dict,count_dict,length_count_dict)
    eval_dict_ngram = evalLineNgram(projectName,seed)

    #IFA
    result_ifa,sorted_eval_dict,ifa_result_dict = getIFA(eval_dict)
    result_ifa_ngram,sorted_eval_dict_ngram,ifa_result_dict_ngram = getIFA(eval_dict_ngram)
    result_top_k = getTopKPrecision(10,sorted_eval_dict)
    result_top_k_ngram = getTopKPrecision(10,sorted_eval_dict_ngram)
#     result_recall_20_loc = getRecallLoc(1,sorted_eval_dict)
#     result_recall_20_loc_ngram = getRecallLoc(1,sorted_eval_dict_ngram)
#     result_effort_20_recall = getEffortRecall(1,sorted_eval_dict)
    result_recall_20_loc,recall_20_loc_number = getRecallLoc(0.2,sorted_eval_dict)
    result_recall_20_loc_ngram,recall_20_loc_number_ngram = getRecallLoc(0.2,sorted_eval_dict_ngram)
    result_effort_20_recall = getEffortRecall(0.2,sorted_eval_dict)
    result_effort_20_recall_ngram = getEffortRecall(0.2,sorted_eval_dict_ngram)
    
    result_mrr,result_mrr_dict = getMRR(sorted_eval_dict)
    result_mrr_ngram,result_mrr_ngram_dict = getMRR(sorted_eval_dict_ngram)
    result_map_list,result_map, result_ap_dict = getMAP(sorted_eval_dict)
    result_map_ngram_list,result_map_ngram, result_ap_ngram_dict = getMAP(sorted_eval_dict_ngram)
    result_percentage_top_k = getPercentageTopK(10,sorted_eval_dict)
    result_percentage_top_k_ngram = getPercentageTopK(10,sorted_eval_dict_ngram)
    #csv
    csv_result = []
    raw_data_result = []
    lime_score_result = []
    fileNameList = getFileNameList(sorted_eval_dict)
    csv_result = generateResultList(csv_result, "our",projectName,fileNameList,result_ifa,result_top_k,result_recall_20_loc,result_effort_20_recall,result_mrr,result_map_list)
    csv_result = generateResultList(csv_result, "ngram",projectName,fileNameList,result_ifa_ngram,result_top_k_ngram,result_recall_20_loc_ngram,result_effort_20_recall_ngram,result_mrr_ngram,result_map_ngram_list)
    print("project:", projectName, "totalFiles:",len(fileNameList))
    
    top_k_result = [{'Technique':"our", 'Datasets':projectName, 'Measure':'Top10Accuracy(File)', 'Value':result_percentage_top_k},
            {'Technique':"ngram", 'Datasets':projectName, 'Measure':'Top10Accuracy(File)', 'Value':result_percentage_top_k_ngram}
           ]
    
    raw_data_result = generateRawDataCsv(raw_data_result,projectName,eval_dict_sum,eval_dict,eval_dict_ngram)
    lime_score_result = generateLimeScoreDataCsv(lime_score_result,projectName,feature_dict,line_predict_dict)
#     ShowFig
#     showFigure(result_ifa,result_ifa_ngram, "IFA")
#     showSingleFigure(recall_20_loc_number, "How many loc 20%")
#     show20effortFigure(result_top_k,result_top_k_ngram, "Top 10 precision")
#     show20effortFigure(result_recall_20_loc,result_recall_20_loc_ngram, "Recall 20% LOC")
#     show20effortFigure(result_effort_20_recall,result_effort_20_recall_ngram, "Effort 20% Recall")
#     show20effortFigure(result_mrr,result_mrr_ngram, "MRR")
    print("MAP",result_map,result_map_ngram)
    print("top-10 percentage of file have at least one in-line comment",result_percentage_top_k,result_percentage_top_k_ngram)
    return csv_result,top_k_result,raw_data_result,lime_score_result
```
